```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
%matplotlib inline
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
from mlxtend.frequent_patterns import association_rules 
frequent_itemsets_=apriori(df, min_support=0.6)
association_rules(frequent_itemsets_,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets_,metric="lift",min_threshold=0.7)
```

```python
Df = pd.read_csv('Market_Basket_Optimisation.csv', header='infer')
Df
```

```python
transactions = []
for i in range(0, 7500):
    transactions.append([str(Df.values[i,j]) for j in range(0, 20)])
```

```python
#print(transactions)
```

```python
te=TransactionEncoder()
te_ary=te.fit(transactions).transform(transactions)    #Apply one-hot-encoding on our dataset
data=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
data
```

```python
from mlxtend.frequent_patterns import apriori
apriori(data, min_support=0.1)
```

```python
from mlxtend.frequent_patterns import association_rules 
frequent_itemsets=apriori(data, min_support=0.1)
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1)
```
